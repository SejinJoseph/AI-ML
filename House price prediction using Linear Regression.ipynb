```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/data.csv')
df
```

```python
df.drop("date",axis=1,inplace=True)
df.drop("view",axis=1,inplace=True)
df.drop("waterfront",axis=1,inplace=True)
df.drop("sqft_basement",axis=1,inplace=True)
df.drop("sqft_above",axis=1,inplace=True)
df.drop("statezip",axis=1,inplace=True)
df.drop("street",axis=1,inplace=True)
df.drop("country",axis=1,inplace=True)
df
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['city']=le.fit_transform(df['city'])
df
```

```python
le.classes_
```

```python
# split the data into training and test sets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Define features (x) and target (y)
y = df[['city','price']]
x = df['price']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

```

```python
lr.fit(X_train.to_frame(), y_train)
```

```python
y_pre=lr.predict(X_test.to_frame())
```

```python
y_pre
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pre)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pre)
plt
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(y_test['city'], y_test['price'], color='b', label='Actual Data')

# Sort the predicted city values and get corresponding predicted price values for plotting the regression line
sorted_city_pred = np.sort(y_pre[:, 0])
sorted_price_pred = y_pre[:, 1][np.argsort(y_pre[:, 0])]

plt.plot(sorted_city_pred, sorted_price_pred, color='r', label='Regression Line')
plt.xlabel('city')
plt.ylabel('price')
plt.legend()
plt.title('Linear Regression: Price vs. city')
plt.show()
```

### Regression Metrics

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# R-squared (R2) score
r2 = r2_score(y_test, y_pre)
print(f"R-squared (R2) Score: {r2:.4f}")

# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pre)
print(f"Mean Absolute Error (MAE): {mae:.2f}")

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pre)
print(f"Mean Squared Error (MSE): {mse:.2f}")
```
